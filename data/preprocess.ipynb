```python
import pandas as pd
```

```python
# 데이터셋 불러오기

meth = pd.read_csv('./crawling/meth.csv')
marijuana = pd.read_csv('./crawling/marijuana.csv')
ecstasy = pd.read_csv('./crawling/ecstasy.csv')
yaba = pd.read_csv('./crawling/yaba.csv')
ketamine = pd.read_csv('./crawling/ketamine.csv')
propofol = pd.read_csv('./crawling/propofol.csv')
cocaine = pd.read_csv('./crawling/cocaine.csv')
lsd = pd.read_csv('./crawling/lsd.csv')
ghb = pd.read_csv('./crawling/ghb.csv')
fentanyl = pd.read_csv('./crawling/fentanyl.csv')
```

```python
# 데이터 프레임 합치기

rawdata = meth.append(marijuana).append(ecstasy).append(yaba).append(ketamine).append(propofol).append(cocaine).append(lsd).append(ghb).append(fentanyl)
```

```python
# 데이터 프레임을 리스트 형태로 변환

rawdata = rawdata['판례원본'].tolist()
```

```python
len(rawdata)
```

```python
# 전처리한 데이터를 넣을 빈 리스트 만들기

data = [[] for row in range(len(rawdata))]
```

```python
# 데이터 전처리

for i in range(len(rawdata)):
    
    # 필요한 부분의 인덱스를 변수로 선언
    
    numberStart = rawdata[i].find('선고')
    numberEnd = rawdata[i].find('판결')
    sentence = rawdata[i].find('\n', 3)
    factStart = rawdata[i].find('법령의 적용')
    factEnd = factStart + 10 + rawdata[i][factStart+10:].find('1.')
    
    caseNumber = rawdata[i][numberStart+3:numberEnd-1] # 사건번호
    drugs = rawdata[i][factStart+10:factEnd] # 검색 외 약물
    resultType = rawdata[i][3:sentence] # 선고 형종
    result = rawdata[i][3:sentence] # 선고 형량
    facts = rawdata[i][factStart+10:factEnd] # 범죄 사실 내용
    
    ### 새 리스트에 데이터 입력
    
    
    # 사건번호
    
    data[i].append(caseNumber)
    
    # 취급마약
    
    if factStart == -1:
        data[i].append('-')
    else:
        drugList = []
        
        if '필로폰' in drugs:
            drugList.append('필로폰')
        elif '암페타민' in drugs:
            drugList.append('필로폰')
        elif '메스암페타민' in drugs:
            drugList.append('필로폰')
        elif '메트암페타민' in drugs:
            drugList.append('필로폰')
        if '대마' in drugs:
            drugList.append('대마')
        elif '해시시' in drugs:
            drugList.append('대마')
        if '엑스터시' in drugs:
            drugList.append('엑스터시')
        elif 'MDMA' in drugs:
            drugList.append('엑스터시')
        if '야바' in drugs:
            drugList.append('야바')
        if '케타민' in drugs:
            drugList.append('케타민')
        if '프로포폴' in drugs:
            drugList.append('프로포폴')
        if '코카인' in drugs:
            drugList.append('코카인')
        if 'LSD' in drugs:
            drugList.append('LSD')
        if 'GHB' in drugs:
            drugList.append('GHB')
        if '펜타닐' in drugs:
            drugList.append('펜타닐')
        
        if drugList == []:
            data[i].append('-')
        else:
            data[i].append(drugList[0])
            drugList = drugList[1:]
    
            for j in drugList:
                data[i][1] += "," + j
    
    # 형량
    
    data[i].append(result)
    
    # 형종
    
    if '집행유예' in resultType:
        data[i].append('집행유예형')
    elif '징역' in resultType:
        data[i].append('징역형')
    elif '벌금' in resultType:
        data[i].append('벌금형')
    else:
        data[i].append('-')
    
    # 형량범위
    
    if data[i][-1] == "징역형":
        if '장기' and '단기' in data[i][-2]:
            data[i].append('-')
        elif '년' in data[i][-2]:
            start = data[i][-2].find('징역')
            end = data[i][-2].find('년')
            tmp = data[i][-2][start+3:end]
            tmp = int(tmp)
            if tmp == 1:
                data[i].append('1년-2년')
            elif tmp == 2:
                data[i].append('2년-3년')
            elif tmp == 3:
                data[i].append('3년-5년')
            elif tmp == 4:
                data[i].append('3년-5년')
            elif tmp == 5:
                data[i].append('5년-7년')
            elif tmp == 6:
                data[i].append('5년-7년')
            elif tmp == 7:
                data[i].append('7년-10년')
            elif tmp == 8:
                data[i].append('7년-10년')
            elif tmp == 9:
                data[i].append('7년-10년')
            else:
                data[i].append('10년이상')
            
        else:
            data[i].append('1년미만')
            
    elif data[i][-1] == "집행유예형":
        if '년' in data[i][-2]:
            start = data[i][-2].find('집행유예')
            end = data[i][-2].rfind('년')
            tmp = data[i][-2][start+5:end]
            tmp = int(tmp)
            if tmp == 1:
                data[i].append('1년')
            elif tmp == 2:
                data[i].append('2년')
            elif tmp == 3:
                data[i].append('3년')
            elif tmp == 4:
                data[i].append('4년')
            else:
                data[i].append('5년이상')
        else:
            data[i].append('1년미만')
    
    elif data[i][-1] == "벌금형":
        start = data[i][-2].find('벌금')
        end = data[i][-2].find('원')
        tmp = data[i][-2][start+3:end]
        tmp = int(tmp.replace(',',''))
        if tmp < 1000000:
            data[i].append('1백만원미만')
        elif tmp >= 1000000 and tmp < 2000000:
            data[i].append('1백만원-2백만원')
        elif tmp >= 2000000 and tmp < 3000000:
            data[i].append('2백만원-3백만원')
        elif tmp >= 3000000 and tmp < 5000000:
            data[i].append('3백만원-5백만원')
        elif tmp >= 5000000 and tmp < 7000000:
            data[i].append('5백만원-7백만원')
        elif tmp >= 7000000 and tmp < 10000000:
            data[i].append('7백만원-1천만원')
        else:
            data[i].append('1천만원이상')
    
    else:
        data[i].append('-')
    
    # 범죄사실
    
    if factStart == -1:
        data[i].append('-')
    else:
        
        factList = []

        if '투약' in facts:
            factList.append('투약')
        if '사용' in facts:
            factList.append('사용')
        if '흡연' in facts:
            factList.append('흡연')
        if '흡입' in facts:
            factList.append('흡입')
        if '소지' in facts:
            factList.append('소지')
        if '소유' in facts:
            factList.append('소유')
        if '매매' in facts:
            factList.append('매매')
        if '매수' in facts:
            factList.append('매수')
        if '매도' in facts:
            factList.append('매도')
        if '알선' in facts:
            factList.append('알선')
        if '제공' in facts:
            factList.append('제공')
        if '수수' in facts:
            factList.append('수수')
        if '수입' in facts:
            factList.append('수입')
        if '수출' in facts:
            factList.append('수출')
        if '제조' in facts:
            factList.append('제조')
        if '조제' in facts:
            factList.append('조제')
        if '운반' in facts:
            factList.append('운반')
        if '재배' in facts:
            factList.append('재배')

        if factList == []:
            data[i].append('-')
        else:
            data[i].append(factList[0])
            factList = factList[1:]

            for j in factList:
                data[i][5] += "," + j

    # 판례원본
    
    data[i].append(rawdata[i][3:])
```

```python
# 제거할 데이터 인덱스를 저장하는 리스트 만들기

tmp = []
```

```python
# 사건 번호가 중복되는 데이터의 인덱스를 tmp에 저장

for i in range(len(data)):
    for j in range(i+1, len(data)):
        if data[i][0] == data[j][0]:
            tmp.append(j)
```

```python
# 결측치 데이터의 인덱스를 tmp에 저장

for i in range(len(data)):
    if data[i][1] == '-': # '취급약물'의 값이 '-' 인 데이터
        tmp.append(i)
    elif data[i][3] == '-': # '형종'의 값이 '-'인 데이터
        tmp.append(i)
    elif data[i][5] == '-': # '범죄사실'의 값이 '-'인 데이터
        tmp.append(i)
```

```python
# 리스트 tmp 내 중복값 제거

tmp = set(tmp)
tmp = list(tmp)
```

```python
# 데이터를 데이터 프레임으로 변환

df = pd.DataFrame(data, columns = ['사건번호', '취급마약', '형량', '형종', '형량범위', '범죄사실', '판례원본'])
```

```python
# 데이터 제거 

df = df.drop(tmp, axis=0)
```

```python
# 인덱스 초기화

df = df.reset_index(drop=True)
```
